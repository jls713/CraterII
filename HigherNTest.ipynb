```python
def load_profiles(ttype='st'):
    fldr = '/data/gaia-eso/jls/m2m/CraterII/vmax_universal/flattening_profiles/'
    fl='2comp_cuspplum_nfw_SP_20_PM_5_ca_10'
    print fldr,fl
    import glob
    fls = glob.glob(fldr+fl+'_flattening_profile_%s*'%ttype)
    times_list = [float(s.split('_')[-1][:-4]) for s in fls]
    fls = np.array(fls)[np.argsort(np.array(times_list))]
    times = np.sort(times_list)
    return times, fls

def load_N_profiles(ttype='st'):
    fldr = '/data/gaia-eso/jls/m2m/CraterII/vmax_universal/flattening_profiles/'
    fl='2comp_cuspplum_nfw_SP_20_PM_5_ca_10_Ntimes10'
    print fldr,fl
    import glob
    fls = glob.glob(fldr+fl+'_flattening_profile_%s*'%ttype)
    times_list = [float(s.split('_')[-1][:-4]) for s in fls]
    fls = np.array(fls)[np.argsort(np.array(times_list))]
    times = np.sort(times_list)
    return times, fls
```

```python
times, fls = load_profiles()
```

```python
times, flsN = load_N_profiles()
```

```python
data = lower_n_properties
mrange=np.zeros(len(fls))
cashape = np.zeros(len(fls))
bashape = np.zeros(len(fls))
epsrange = np.zeros(len(fls))
srange = np.zeros(len(fls))
for II, F in enumerate(fls):
    dd = pd.read_csv(F)
    if II>= len(data):
        continue
    mrange[II] = data['mass_dm_core_init'].values[II]/data['mass_dm_core_init'][0]
    epsrange[II] = data['eps'].values[II]
    bashape[II] = np.mean(dd['ba_st'][dd['radii_st']<.5*data['r_half'][0]])
    cashape[II] = np.mean(dd['ca_st'][dd['radii_st']<.5*data['r_half'][0]])
    srange[II] = data['s'].values[II]
    
data = higher_n_properties
mrangeN=np.zeros(len(flsN))
cashapeN = np.zeros(len(flsN))
bashapeN = np.zeros(len(flsN))
epsrangeN = np.zeros(len(flsN))
srangeN = np.zeros(len(flsN))
for II, F in enumerate(flsN):
    dd = pd.read_csv(F)
    if II>= len(data):
        continue
    mrangeN[II] = data['mass_dm_core_init'].values[II]/data['mass_dm_core_init'][0]
    epsrangeN[II] = data['eps'].values[II]
    bashapeN[II] = np.mean(dd['ba_st'][dd['radii_st']<.5*data['r_half'][0]])
    cashapeN[II] = np.mean(dd['ca_st'][dd['radii_st']<.5*data['r_half'][0]])
    srangeN[II] = data['s'].values[II]
```
